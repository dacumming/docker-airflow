```python
def add_column(engine, table_name, column):
    column_name = column.compile(dialect=engine.dialect)
    column_type = column.type.compile(engine.dialect)
    engine.execute('ALTER TABLE %s ADD COLUMN %s %s' % (table_name, column_name, column_type))
```

```python
track = Column('track', String(90))
bass = Column('bass', String(90))
drums = Column('drums', String(90))
other = Column('other', String(90))
vocals = Column('vocals', String(90))
track_spec = Column('track_spec', String(90))
bass_spec = Column('bass_spec', String(90))
drums_spec = Column('drums_spec', String(90))
other_spec = Column('other_spec', String(90))
vocals_spec = Column('vocals_spec', String(90))
add_column(engine, 'tracks', track)
add_column(engine, 'tracks', bass)
add_column(engine, 'tracks', drums)
add_column(engine, 'tracks', other)
add_column(engine, 'tracks', vocals)
add_column(engine, 'tracks', track_spec)
add_column(engine, 'tracks', bass_spec)
add_column(engine, 'tracks', drums_spec)
add_column(engine, 'tracks', other_spec)
add_column(engine, 'tracks', vocals_spec)
```
